# Train Test Split

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,label,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Build a pipe line
1. Apply the transformation for handling Outlier and skewed Data

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,FunctionTransformer
from sklearn.linear_model import LinearRegression,Ridge,SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,ExtraTreesRegressor
```

**Postive_col contain all the positive col so that i can apply the log_transform and outlier handle fun that i make above**

```python
pos_trans=Pipeline(steps=[
#     ('handle_outier',FunctionTransformer(handle_outlier,validate=True)),
#     ('normal_distrubut',FunctionTransformer(np.log1p,validate=True)),
    ('scale',StandardScaler())
])
```

**Neg_col contain all the neg col so that i can scale them**

```python
neg_trans=Pipeline(steps=[
#     ('handle_outier',FunctionTransformer(handle_outlier,validate=True)),
#     ('handle_outier',FunctionTransformer(np.log1p,validate=True)),
    ('scale',StandardScaler())
])
```

# Build a Transformer

```python
process=ColumnTransformer(transformers=[
    ('transform',pos_trans,pos_col),
    ('transform2',neg_trans,neg_col)
])
```

# Load the model

```python
from sklearn.linear_model import LinearRegression
```

```python
pipe=Pipeline(steps=[
    ('process',process),
    ('model',LinearRegression())
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pre=pipe.predict(x_test)
```

# Calculate the Accuracy

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
score=r2_score(y_test,pre)
error=mean_squared_error(y_test,pre)
```

```python
score
```

```python
error
```

```python
train_model=cross_val_score(pipe,x_train,y_train,cv=5,scoring='r2')
test_model=cross_val_score(pipe,x_test,y_test,cv=5,scoring='r2')
```

```python
print(train_model.mean())
print(test_model.mean())
```

```python
plt.figure(figsize=(16, 6))
sns.regplot(x=y_test, y=pre, scatter_kws={'alpha':0.5})
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (pre)')
plt.title('Regression Plot of Actual vs. Predicted Values')
plt.grid(True)
plt.show()
```
